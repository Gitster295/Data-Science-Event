# 7. Create a bar plot to compare categories of a column.

```python
car_counts = data['CarModel'].value_counts()
plt.figure(figsize=(10, 6))
car_counts.plot(kind='bar', color='#A6D1C3')
plt.title('Number of Rentals per Car Model')
plt.xlabel('Car Model')
plt.ylabel('Number of Rentals')
plt.show()
```

# 8. Generate a heatmap to show correlation among numerical columns.

```python
numerical_data = data.select_dtypes(include=['number'])

corr_matrix = numerical_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

plt.title('Correlation Heatmap of Numerical Columns')
plt.tight_layout()
plt.show()
```

# 9. Create a line plot or time series plot to analyze trends over time.

```python
data['BookingDate'] =pd.to_datetime(data['BookingDate'])
data['ReturnDate'] = pd.to_datetime(data['ReturnDate'])
```

```python
data['BookingDate'] = pd.to_datetime(data['BookingDate'])
data['BookingMonth'] = data['BookingDate'].dt.to_period('M').astype(str)

trend = data.groupby(['BookingMonth', 'CarModel'])['RentalCost'].sum().unstack()

trend.plot(kind='line', figsize=(12, 6), marker='s')
plt.title('Monthly Rental Revenue by Car Model')
plt.xlabel('Month')
plt.ylabel('Total Rental Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```
